```python
import numpy as np
```

```python
# Jumlah simulasi
N = 10000

# Distribusi kapasitas
kapasitas = np.random.normal(loc=80.0, scale=10, size=N)

# Distribusi debit
debit = np.random.normal(loc=70, scale=15, size=N)

# Fungsi kegagalan
g = kapasitas - debit
```

```python
#B
B = (80-70)/np.sqrt(10**2+15**2)

# Probabilitas kegagalan
Pf = np.sum(g < 0) / N

# Keandalan
R = 1 - Pf

print(B)
print(Pf)
print(R)
```
