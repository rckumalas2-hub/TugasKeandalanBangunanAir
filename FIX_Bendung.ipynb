```python
import numpy as np
```

```python
# Jumlah simulasi
N = 10000

# Distribusi kapasitas pelimpah
kapasitas = np.random.normal(loc=1200.0, scale=150, size=N)

# Distribusi debit banjir
debit = np.random.normal(loc=1000, scale=250, size=N)

# Fungsi kegagalan
g = kapasitas - debit
```

```python
#B
B = (1200-1000)/np.sqrt(150**2+250**2)

# Probabilitas kegagalan
Pf = np.sum(g < 0) / N

# Keandalan
R = 1 - Pf

print(B)
print(Pf)
print(R)
```

```python
#Hitung tingkat keamanan!
##  Tingkat keamanan = keandalan,
##  maka nilainya adalah 74.51%
```
