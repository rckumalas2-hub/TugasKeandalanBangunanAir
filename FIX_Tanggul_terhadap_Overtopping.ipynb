```python
import numpy as np
```

```python
# Jumlah simulasi
N = 10000

# Distribusi elevasi tanggul
crest = np.random.normal(loc=5.0, scale=0.2, size=N)

# Distribusi muka air banjir
water = np.random.normal(loc=4.2, scale=0.5, size=N)

# Fungsi kegagalan
g = crest - water
```

```python
#B
B = (5-4.2)/np.sqrt(0.2**2+0.5**2)

# Probabilitas kegagalan
Pf = np.sum(g < 0) / N

# Keandalan
R = 1 - Pf

print(B)
print(Pf)
print(R)
```
